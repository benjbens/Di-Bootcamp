```python
url = 'https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv'
cars_df = pd.read_csv(url).dropna(subset=['Manufacturer', 'Price'])
```

```python
cars_df.head()
```

---
Use Target encoding to create an presentative column to a target column (in this case the Manufacturer) 

```python
means = cars_df[['Manufacturer', 'Price']].groupby('Manufacturer').mean()
means
```

```python
cars_df.info()
```

```python
cars_df['Manufacturer_transformed'] = cars_df['Manufacturer'].apply(lambda x: means.loc[x, "Price"])
```

```python
cars_df.head()
```

---
Label encoding of the Make column

```python
make_column = cars_df.Make
```

```python
make_column.head(2)
```

```python
d = ['a','b','c']
indx_vals = list(enumerate(d))
indx_vals
```

```python
# List comprehension - how to create a list in one line of code
[x for x in range(10)]
```

```python
# Create the mapping dictionary with Key as the String that we want to encode and the
# Value as the number we want to associate it with. 
# Dictionary comprehension - how to create a dictionary in one line of code
make_dict = {v: idx for idx, v in enumerate(make_column)}
```

```python
# Use the column_name.map function to perform the mapping. 
encoded_make = make_column.map(make_dict)
```

```python
make_column['encoded'] = encoded_make
```

```python
make_column.head()
```

```python
cars_df['Make_encoded'] = make_column['encoded']
```

```python
cars_df.head()
```

```python
cars_df.columns
```

--- 
Target encoding of the Model

```python
cars_df['Model'].fillna('Nan', inplace = True)
```

```python
cars_df['Model'].isnull().values.any()
```

```python
means_price = cars_df[['Model', 'Price']].groupby('Model').mean()
means_horsepower = cars_df[['Model', 'Horsepower']].groupby('Model').mean()
```

```python
means_price.shape
```

```python
means_horsepower.shape
```

```python
means_price
```

```python
means_horsepower
```

```python
price_horsepower_combined = means_price['Price'] + means_horsepower['Horsepower']
```

```python
price_horsepower_combined.head()
```

```python
price_horsepower_combined.index
```

```python
price_horsepower_combined['100']
```

```python
cars_df['Model_encoded'] = cars_df['Model'].apply(lambda x: price_horsepower_combined[x])
```

```python
cars_df.head()
```
